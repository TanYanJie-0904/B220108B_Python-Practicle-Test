```python
def display_hangman(incorrect_guesses):
    stages = [
        """
           -----
           |   |
               |
               |
               |
               |
        --------
        """,
        """
           -----
           |   |
           O   |
               |
               |
               |
        --------
        """,
        """
           -----
           |   |
           O   |
           |   |
               |
               |
        --------
        """,
        """
           -----
           |   |
           O   |
          /|   |
               |
               |
        --------
        """,
        """
           -----
           |   |
           O   |
          /|\\  |
               |
               |
        --------
        """,
        """
           -----
           |   |
           O   |
          /|\\  |
          /    |
               |
        --------
        """,
        """
           -----
           |   |
           O   |
          /|\\  |
          / \\  |
               |
        --------
        """
    ]
    print(stages[incorrect_guesses])


```

```python
import random

def hangman():
    word_list = ["PYTHONIC", "PROGRAMMING", "COMPUTER", "DEVELOPER", "FUNCTIONAL", "SOFTWARE"]
    chosen_word = random.choice(word_list).upper()
    hidden_word = ["_"] * len(chosen_word)
    guessed_letters = []
    max_incorrect_guesses = 6
    incorrect_guesses = 0

    print("Welcome to Hangman!")
    
    while incorrect_guesses < max_incorrect_guesses and "_" in hidden_word:
        print("\nCurrent word: ", " ".join(hidden_word))
        display_hangman(incorrect_guesses)
        print(f"Incorrect guesses remaining: {max_incorrect_guesses - incorrect_guesses}")
        guess = input("Guess a letter: ").upper()

        if guess in guessed_letters:
            print(f"You've already guessed '{guess}'. Try another letter.")
        elif guess in chosen_word:
            guessed_letters.append(guess)
            for i in range(len(chosen_word)):
                if chosen_word[i] == guess:
                    hidden_word[i] = guess
            print(f"Good guess! '{guess}' is in the word.")
        else:
            guessed_letters.append(guess)
            incorrect_guesses += 1
            print(f"Sorry, '{guess}' is not in the word.")
        
    if "_" not in hidden_word:
        print("\nCongratulations! You've guessed the word:", " ".join(hidden_word))
    else:
        print("\nYou've run out of guesses. The word was:", chosen_word)
        display_hangman(incorrect_guesses)
```

```python
hangman()
```
